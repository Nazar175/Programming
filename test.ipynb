```python
class StringReverser:
    def __init__(self, text: str) -> None:
        self.text = text
        self._reversed: str | None = None

    def reverse(self) -> str:
        self._reversed = self.text[::-1]
        return self._reversed

    @property
    def reversed_string(self) -> str:
        if self._reversed is None:
            self.reverse()
        print(f"Рядок у зворотному порядку: {self._reversed}")
        return self._reversed

    def __str__(self) -> str:
        return "Виконав: Гучко Назар, Група: КН-33"


texts = [
    "Це тестова стрічка",
    "Програмування на Python",
    "Залік з ООП"
]

for text in texts:
    reverser = StringReverser(text)
    print(reverser)
    print(f"Початковий рядок: {text}")
    _ = reverser.reversed_string
    print("-" * 40)
```

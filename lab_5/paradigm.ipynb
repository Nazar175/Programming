# –î–æ–¥–∞–π–Ω–æ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö —á–∏—Å–µ–ª —Ç–∞ –≤–∏–∫–ª–∏–∫–∞–Ω–æ –º–µ—Ç–æ–¥ deposit —Ç–∞ withdraw.

```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f"–ü–æ–ø–æ–≤–Ω–µ–Ω–Ω—è: +{amount} –≥—Ä–Ω")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"–ó–Ω—è—Ç—Ç—è: -{amount} –≥—Ä–Ω")
            return amount
        else:
            print(f"–°–ø—Ä–æ–±–∞ –∑–Ω—è—Ç—Ç—è {amount} –≥—Ä–Ω ‚Äî –Ω–µ–¥–æ—Å—Ç–∞—Ç–Ω—å–æ –∫–æ—à—Ç—ñ–≤")
            return "Insufficient funds"

    def get_balance(self):
        return self.__balance


account = BankAccount("Bohdan", 1000)

for i in range(10):
    operation = random.choice(["deposit", "withdraw"])
    amount = random.randint(100, 500)

    if operation == "deposit":
        account.deposit(amount)
    else:
        account.withdraw(amount)

print("\n–ö—ñ–Ω—Ü–µ–≤–∏–π –±–∞–ª–∞–Ω—Å:", account.get_balance(), "–≥—Ä–Ω")
```

# –°—Ç–≤–æ—Ä–µ–Ω–æ —â–µ –æ–¥–∏–Ω –º–µ—Ç–æ–¥ —É –∫–ª–∞—Å—ñ Vehicle —Ç–∞ –≤–∏–∫–ª–∏–∫–∞–Ω–æ –π–æ–≥–æ –∑ –æ–±—î–∫—Ç–∞ –∫–ª–∞—Å—É Car

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"
    
    def start_engine(self):
        return f"{self.brand} {self.model} engine started."


class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"


car = Car("Toyota", "Camry", 5)

print(car.display_info())
print(car.start_engine())
```

# –°—Ç–≤–æ—Ä–µ–Ω–æ –∫–ª–∞—Å Fish —É —è–∫–æ–≥–æ –Ω–µ –±—É–¥–µ –º–µ—Ç–æ–¥–∞ speak

```python
class Animal:
    def speak(self):
        pass


class Dog(Animal):
    def speak(self):
        return "Woof!"


class Cat(Animal):
    def speak(self):
        return "Meow!"


class Fish(Animal):
    pass


animals = [Dog(), Cat(), Fish()]

for animal in animals:
    print(animal.speak())
```

# –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –ø–∞—Ä–∞–¥–∏–≥–º –¥–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –ø—Ä–æ—Å—Ç–æ—ó –≥—Ä–∏

```python
from abc import ABC, abstractmethod
from random import randint, choice


class Item(ABC):
    def __init__(self, name: str, health=500):
        self.name = name
        self.health = health
        self._buff = 0

    @abstractmethod
    def attack(self, another_item):
        pass

    def apply_buff(self):
        self._buff += 10
        return f"{self.name} –æ—Ç—Ä–∏–º—É—î –ø—ñ–¥—Å–∏–ª–µ–Ω–Ω—è (+10 –¥–æ –Ω–∞—Å—Ç—É–ø–Ω–æ—ó –∞—Ç–∞–∫–∏)"


class Sword(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name)
        self.__attack_power = attack_power
        self._sharp = 0

    def attack(self, another_item):
        current_attack = self.__attack_power + self._sharp + randint(0, 10) + self._buff
        self._buff = 0
        another_item.health -= current_attack
        return f"‚öî –ú–µ—á {self.name} –∑–∞–≤–¥–∞—î {current_attack} —à–∫–æ–¥–∏"

    def sharpening(self):
        self._sharp += 1
        return f"{self.name} –∑–∞–≥–æ—Å—Ç—Ä–µ–Ω–æ. Sharp = {self._sharp}"


class Axe(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name)
        self.__attack_power = attack_power

    def attack(self, another_item):
        current_attack = self.__attack_power + randint(0, 20) + self._buff
        self._buff = 0
        another_item.health -= current_attack
        return f"ü™ì –°–æ–∫–∏—Ä–∞ {self.name} –∑–∞–≤–¥–∞—î {current_attack} —à–∫–æ–¥–∏"


class Bow(Item):
    def __init__(self, name, attack_power: int, range_power: int):
        super().__init__(name)
        self.__attack_power = attack_power
        self.range_power = range_power

    def attack(self, another_item):
        current_attack = self.__attack_power + randint(5, 15) + self.range_power + self._buff
        self._buff = 0
        another_item.health -= current_attack
        return f"üèπ –õ—É–∫ {self.name} –∑–∞–≤–¥–∞—î {current_attack} —à–∫–æ–¥–∏"

    def reload(self):
        self.range_power += 1
        return f"{self.name} –ø—ñ–¥–≤–∏—â—É—î –¥–∞–ª—å–Ω—ñ—Å—Ç—å. Range = {self.range_power}"


weapons = [
    Sword("–ï—Å–∫–∞–ª—ñ–±—É—Ä", 100),
    Axe("–õ–µ–≤—ñ–∞—Ñ–∞–Ω", 95),
    Bow("–ï–ª—å—Ñ—ñ–π—Å—å–∫–∏–π –ª—É–∫", 85, 5)
]

player = choice(weapons)
enemy = choice([w for w in weapons if w != player])

print(f"üéÆ –í–∏ –≥—Ä–∞—î—Ç–µ –∑–∞: {player.name}")
print(f"ü§ñ –ü—Ä–æ—Ç–∏–≤–Ω–∏–∫: {enemy.name}")
print("-" * 40)


turn = 1
while player.health > 0 and enemy.health > 0:
    print(f"\n–•—ñ–¥ {turn}")
    print(f"–í–∞—à–µ HP: {player.health} | HP –≤–æ—Ä–æ–≥–∞: {enemy.health}")
    print("1 - –ê—Ç–∞–∫—É–≤–∞—Ç–∏")
    print("2 - –ü—ñ–¥—Å–∏–ª–µ–Ω–Ω—è")
    
    if isinstance(player, Sword):
        print("3 - –ó–∞–≥–æ—Å—Ç—Ä–∏—Ç–∏ –º–µ—á")
    if isinstance(player, Bow):
        print("3 - –ù–∞—Ç—è–≥–Ω—É—Ç–∏ —Ç—è—Ç–∏–≤—É (reload)")

    choice_input = input("–û–±–µ—Ä—ñ—Ç—å –¥—ñ—é: ")

    if choice_input == "1":
        print(player.attack(enemy))
    elif choice_input == "2":
        print(player.apply_buff())
    elif choice_input == "3":
        if isinstance(player, Sword):
            print(player.sharpening())
        elif isinstance(player, Bow):
            print(player.reload())
        else:
            print("–ù–µ–º–∞—î —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–æ—ó –¥—ñ—ó")
    else:
        print("–ù–µ–≤—ñ—Ä–Ω–∏–π –≤–∏–±—ñ—Ä")

    if enemy.health <= 0:
        print(f"\nüèÜ –ü–µ—Ä–µ–º–æ–≥–∞ –∑–∞ {player.name}")
        break

    enemy_action = randint(1, 2)
    if enemy_action == 1:
        print(enemy.attack(player))
    else:
        print(enemy.apply_buff())

    if player.health <= 0:
        print(f"\nüíÄ –ü–µ—Ä–µ–º–æ–≥–∞ –∑–∞ {enemy.name}")
        break

    turn += 1
```
